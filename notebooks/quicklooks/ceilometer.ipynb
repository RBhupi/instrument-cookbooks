# Ceilometer Quicklooks

## Code to Generate Plots

```python
import act
import glob
import matplotlib.pyplot as plt

files = glob.glob('../../data/ceilometer/*')

ds = act.io.read_netcdf(files)

ds.attrs['datastream'] = 'anl.ceilometer.a1'

variables=['p_pol', 'x_pol', 'linear_depol_ratio']

for variable in variables:
    display = act.plotting.TimeSeriesDisplay(ds, figsize=(8,4))
    display.plot(variable, cmap='Spectral_r')
    plt.show()
    plt.close()
```
